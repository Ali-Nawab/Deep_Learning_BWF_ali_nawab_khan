```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import glob
```

```python
# list all the csv files in the directory.
file_list = glob.glob('*.csv')

# create the empty list to hold the dataframe.
df_list = []

# loop through the file list and append each file in df_list.
for filename in file_list:
    df_list.append(pd.read_csv(filename))
    
# concatenate all the dataframe in the list to one single dataframe.
us_census = pd.concat(df_list, ignore_index=True)
```

```python
us_census['Pacific'] = us_census['Pacific'].str.replace('%', '')
```

```python
us_census['Pacific'] = pd.to_numeric(us_census['Pacific'])
```

```python
us_census['Pacific'] = us_census['Pacific'].fillna(us_census['Pacific'].mean())
```

```python
us_census['Income'] = us_census['Income'].str.replace('$', '')
```

```python
us_census['Income'] = pd.to_numeric(us_census['Income'])
```

```python
for column in ['Hispanic', 'White', 'Black', 'Native', 'Asian']:
    us_census[column] = us_census[column].str.replace('%', '')
    us_census[column] = pd.to_numeric(us_census[column])
```

```python
us_census['Income'] = us_census['Income'].round(3)
```

```python
for column in ['Hispanic', 'White', 'Black', 'Native', 'Asian', 'Pacific']:
    us_census[column] = us_census[column].round(3)
```

```python
# Split GenderPop column into Men and Women columns
us_census[['Men', 'Women']] = us_census['GenderPop'].str.split('_', expand=True)
```

```python
# Remove non-numeric characters from Men and Women columns and convert to numerical data types
us_census['Men'] = pd.to_numeric(us_census['Men'].str.replace('[^0-9]', ''))
us_census['Women'] = pd.to_numeric(us_census['Women'].str.replace('[^0-9]', ''))
```
